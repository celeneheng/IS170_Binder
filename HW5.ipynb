```python
#set up the libraries
# we use apriori for SUPPORT and association_rules for CONFIDENCE, LIFT and CONVICTION
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Since it's missing a column header, so we create new column header name
df=pd.read_csv('AUTOZONEDataSet.csv', names=['autoparts'], sep=',')
df.head()
```

```python
#data size
df.shape
```

```python
#the following code converts a pandas dataframe into a list of lists.
data=list(df["autoparts"].apply(lambda x:x.split(",")))
data
```

```python
#Let's transform the list with one-hot encoding
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)
df=df.replace(True,1)
df=df.replace(False,0)
df
```

```python
#Set a threshold value for the support value and calculate the support value
df=apriori(df,min_support=0.2, use_colnames=True)
df
```

Support is used to show the frequency of how many times the items was purchased out of all the purchases

```python
#Here is the CONFIDENCE (Try different min_threshold value to explore this more)
df_ar=association_rules(df,metric="confidence", min_threshold=0.3)
df_ar
```

The first and second rows (Coolant -> Car Scents) show CONFIDENCE .57 and .31 which add up to .89 which means the two items are bought togetehr 89%. The third row (wiper pluid -> car soap show 0.57 which means the two items are bought together only 35% (or when wiper fluid is bought only 57% chance car soap is also bought).

confidence shows the probablity of and antecedent and consequent occuring together.

```python
#Here is LIFT (try different min_threshold value to explore more) 
df_ar=association_rules(df,metric="lift", min_threshold=0.8)
df_ar
```

The two rows (Coolant & car scents) show LIFT 1.8 which is a positive correlation between the two. The rows 2 and 3 (Wiper fluid -> car soap show LIFT 4.5 which is high positive correlation between the two. 

Lift is about the correlation between the antecedant and consquent

```python
#Let's view our interpretation values using the Association rule function
df_ar=association_rules(df,metric="conviction", min_threshold=0.8)
df_ar
```

CONVICTION is same as LIFT except CONVICTION also gives the directed measure. Meaning it gives the correlation when they are purchased seperately. If you look at all the rows the convictions are close to 1 none of them are infinity therfore there are no correlations in this data set.

```python
#Since it's missing a header we'll create one, the new column header 'Products' 
df=pd.read_csv('fastfood_dataset.csv', names=['fast_food'])
df.head() 
```

```python
df.shape
```

```python
#the following code converts a pandas dataframe into a list of lists.
data=list(df["fast_food"].apply(lambda x:x.split(",")))
data
```

```python
#Let's transform the list with one-hot encoding (make sure you understand the one-hot encoding concept first)
#Each column represents an item and each row represent a record or a transaction. 
#Initially, if that particular item occurs then it's labeled TRUE, if not then FALSE.
#After that, TRUE is converted to 1 and FALSE to 0 because computer loves numeric values.

from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True to 0 & 1 for easy readability
df=df.replace(False,0)
df=df.replace(True,1)

df
                
```

```python
#We'll try 0.2 for the SUPPORT threshold value (but you can set it to any value you want).
# The parameter (use_colnames=True) is used to convert the integer back to its respective name.
# apriori is only used with SUPPORT not the others.

df=apriori(df,min_support=0.2,use_colnames=True)
df
```

With the SUPPORT values above, you can try and match your own manual calculation. You notice how some are grouped with two. Not only single item, but also group of two or more are tested. Any item or group of items didn't make 0.2 then they are not listed. French fries were sold 100% of the time with every transaction. Hotdogs were also regularly sold about 60% of the transactions.

```python
#For the CONFIDENCE parameter, we use the 'association_rules' function below.
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

From above result, shows that all the rows have a confidence of 1 meaning that the pairing occur 100% of the time together.

```python
#Here is the LIFT parameter.
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

From above result, the LIFT  tells us there is not a correlation when apples and french fries are purchased. There was a slight positve correlation between apples and french fries. There was a high positive correlation between apples and chili fries. 

```python
#Here is the CONVICTION parameter. 
df_ar=association_rules(df,metric="conviction", min_threshold=1.0)
df_ar
```

From above result, the CONVICTION tells us that there is positive correlation between apples and french fries and apples and chili fries.
all other rows did not show a correlation.
