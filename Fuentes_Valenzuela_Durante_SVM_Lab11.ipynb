```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
```

```python
mydata=pd.read_csv("dementia_dataset.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
mydata
```

```python
le=LabelEncoder()
```

```python
label=le.fit_transform(mydata["Group"])
label
```

```python
mydata.drop("Group", axis=1, inplace=True)
mydata["Group"] = label
mydata
```

```python
label=le.fit_transform(mydata["M/F"])
label
```

```python
mydata.drop("M/F", axis=1, inplace=True)
mydata["M/F"] = label
mydata
```

```python
label=le.fit_transform(mydata["Hand"])
label
```

```python
mydata.drop("Hand", axis=1, inplace=True)
mydata["Hand"] = label
mydata
```

```python
mydata
```

```python
print(mydata.isnull().sum())
```

```python
mydata.fillna(value=0, inplace=True)
mydata
```

```python
mydata["SES"]=mydata["SES"].fillna(value=0)
mydata
```

```python

mydata["MMSE"]=mydata["MMSE"].fillna(value=0)
mydata
```

```python
mydata
```

```python
#the x data is saved without 'target' column to make only features dataframe
x=mydata.drop('Group',axis=1)
```

```python
#the y data is saved only with 'target' column
y=mydata['Group']
```

```python
#create a model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
#kernel specifying
clf=svm.SVC(kernel='linear')

#train and fit the model
clf.fit(x_train,y_train)
```

```python
#predict the response
y_pred=clf.predict(x_test)
```

```python
#model accuracy
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
precision = metrics.precision_score(y_test, y_pred, average='macro')
recall = metrics.recall_score(y_test, y_pred, average='macro')

print("Precision:", precision)
print("Recall:", recall)
```
