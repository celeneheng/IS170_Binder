```python
df = pandas.read_csv('healthcare_data-2.csv') 
```

```python
df.info() #information about a DataFrame
```

```python
df.dropna(how='all') #removed any row with a missing value
df
```

```python
df[['Insulin','SkinThickness']]=df[['Insulin','SkinThickness']].fillna(value=df[['Insulin','SkinThickness']].mean())
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder( )

label = le.fit_transform(df['Outcome'])

label
```

```python
df.drop("Outcome", axis= 1, inplace=True)

df["Outcome"] = label

df
```

```python
from sklearn import preprocessing

df_scaled = preprocessing.scale(df)

df_scaled=pandas.DataFrame(df_scaled,columns=df.columns)

df_scaled['Outcome']=df['Outcome']
df=df_scaled
```

```python
old_cols=df.columns.values
new_cols=['Insulin','Age','BMI','Glucose']
df=df.reindex(columns=new_cols)
df
```
