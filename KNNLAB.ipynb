# Make sure you are cleared with the feature engineering BEFORE the next step (model building). Double check for the missing data values, normalization, string to numeric conversion, etc....

```python
#Import train_test_split function
from sklearn.model_selection import train_test_split

#split dataset into training set and test set
#70% training and 30% test
X_train, X_test, y_train, y_test=train_test_split(wine.data, wine.target, test_size=0.3) 
```

```python
#Generating a model with K=5
#Import k-nearest neighbors classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn=KNeighborsClassifier(n_neighbors=5)

#Train the model using the training sets
knn.fit(X_train,y_train)

#Predict the response for test dataset
y_pred=knn.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

#Model accuracy, how often is the classifier correct?
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

###  From the result above, the classification rate is about 70%. You may change the K value to see what other rates you get.

## We look at the correlation of the other features in comparison to quality to see how impactful each of the features were in judging its value. That means we can safely remove 'residual sugar', 'free sulfur dioxide', and 'pH' as they have a weak correlation. 

```python
new_cols =  ['fixed acidity', 'volatile acidity', 'citric acid', 'chlorides', 'total sulfur dioxide', 'density', 'sulphates', 'alcohol', 'quality']
df= df.reindex(columns = new_cols)
df.head()
```

```python
bins = [0, 10, 12, 15]
labels = ["low","median","high"]
df['alcohol_label'] = pd.cut(df['alcohol'], bins=bins, labels=labels)
df.drop('alcohol',axis =1, inplace = True)
df.head()
```

```python
bins = [0, 4, 6, 10]
labels = ["poor","normal","excellent"]
df['quality_label'] = pd.cut(df['quality'], bins=bins, labels=labels)
df.drop('quality',axis =1, inplace = True)
df.head()
```

```python
df['alcohol_label'].value_counts()
```

```python
df_wine = pd.get_dummies(df, columns=['alcohol_label'], drop_first=True)
df_wine.head()
```

```python
result = df_wine['quality_label']
df_wine.drop(['quality_label'], axis=1, inplace=True)
print(df_wine.shape, result.shape)
```

```python
# use 70% of the data for training and 30% for testing
X_train, X_test, Y_train, Y_test = train_test_split(df_wine, result, test_size=0.30, random_state=11)
```

```python
# For KNN our dataset must have to standardised.
# No need standardised quality_label as it is the result column

scaler = StandardScaler()
scaler.fit(df_wine)
scaled_features = scaler.transform(df_wine)
df_wine_sc = pd.DataFrame(scaled_features, columns=df_wine.columns)
```

```python
# use 70% of the data for training and 30% for testing
X_train_sc, X_test_sc, y_train_sc, y_test_sc = train_test_split(df_wine_sc, result, test_size=0.30, random_state=11)
```

```python
# convert dataframe to nd numpy array
X_train_sc = X_train_sc.to_numpy()
y_train_sc = y_train_sc.to_numpy()
```

```python
def apply_knn(neigh, weight='uniform'):
    knn = KNeighborsClassifier(n_neighbors=neigh, weights=weight)
    knn.fit(X_train_sc,y_train_sc)
    pred_knn = knn.predict(X_test_sc)
    return pred_knn
```

```python
pred_knn_for_20 = apply_knn(20)
print('Accuracy of model at K=20 is', accuracy_score(y_test_sc, pred_knn_for_20))
```

## The results shows a 82% accuracy which is  nearly optimal score for the data. We can measure the alcohol contents to accurately predict the quality of the wine.
